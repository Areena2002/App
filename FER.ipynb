```python
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace

img = cv2.imread('Girl.jpeg')

predictions = DeepFace.analyze(img)

dominant_emotion = predictions[0]['dominant_emotion']

faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces = faceCascade.detectMultiScale(gray, 1.1, 4)

for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
    cv2.putText(img, dominant_emotion, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Face Detection with Emotion")
plt.axis("off")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace

# Load the new image
img1 = cv2.imread('Boy1.jpeg')

# Analyze the image, allowing for potential face detection failures
predictions = DeepFace.analyze(img1, enforce_detection=False)

# Extract the dominant emotion from the analysis results
dominant_emotion = predictions[0]['dominant_emotion']

# Initialize the face cascade classifier
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Convert the image to grayscale for face detection
gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

# Detect faces in the grayscale image
faces = faceCascade.detectMultiScale(gray, 1.1, 4)

# Draw rectangles around detected faces and display the dominant emotion
for (x, y, w, h) in faces:
    cv2.rectangle(img1, (x, y), (x + w, y + h), (0, 255, 0), 2)
    cv2.putText(img1, dominant_emotion, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

# Display the image with face detection and emotion label
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title("Face Detection with Emotion")
plt.axis("off")
plt.show()
```

```python
import cv2  # pip install opencv-python  # pip install opencv-contrib-python for full package
from deepface import DeepFace  # pip install deepface

faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Try different camera indices, including -1 for automatic selection
for camera_index in range(-1, 10):  # Check indices -1, 0 to 9
    cap = cv2.VideoCapture(camera_index)
    if cap.isOpened():
        break  # Exit the loop if a camera is found

if not cap.isOpened():
    raise IOError("Cannot open webcam. Please check your webcam connection, drivers, and permissions.")

while True:
    ret, frame = cap.read()
    result = DeepFace.analyze(frame, actions=['emotion'])
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = faceCascade.detectMultiScale(gray, 1.1, 4)

    # Draw a rectangle around the faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)  # Indented this line to be part of the for loop

    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(frame, result['dominant_emotion'], (50, 50), font, 3, (0, 0, 255), 2, cv2.LINE_4)
    cv2.imshow('Original video', frame)

    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
